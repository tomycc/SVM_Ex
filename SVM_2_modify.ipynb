```python
import pandas as pd 
from sklearn.datasets import load_iris
iris = load_iris()
```

```python
dir(iris)
```

```python
iris.feature_names
```

```python
df = pd.DataFrame(iris.data, columns=iris.feature_names)
```

```python
df.head()
```

```python
df["target"] = iris.target
```

```python
df.head()
```

```python
iris.target_names
```

```python
df[df.target==1].head()
```

```python
df[df.target==2].head()
```

```python
df["flower_name"] = df.target.apply(lambda x: iris.target_names[x])
df.head()
```

```python
from matplotlib import pyplot as plt
```

```python
d={}
for x in range(len(iris.target_names)):
    d["df{0}".format(x)]=df[df.target==x]
```

```python
d
```

```python
color = ["green", "blue", "red"]
```

```python
marker = ["+", "o", "."]
```

```python
plt.xlabel("sepal length (cm)")
plt.ylabel("sepal width (cm)")
for x in range(len(iris.target_names)):
    plt.scatter(d[f"df{x}"]["sepal length (cm)"], d[f"df{x}"]["sepal width (cm)"], color = color[x], marker = marker[x])
```

```python
plt.xlabel("petal length (cm)")
plt.ylabel("petal width (cm)")
for x in range(len(iris.target_names)):
    plt.scatter(d[f"df{x}"]["petal length (cm)"], d[f"df{x}"]["petal width (cm)"], color = color[x], marker = marker[x])
```

```python
from sklearn.model_selection import train_test_split
```

```python
df.head()
```

```python
X = df.drop(['target','flower_name'], axis = 'columns')
```

```python
y = df.target
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
```

```python
len(X_train)
```

```python
len(X_test)
```

```python
from sklearn.svm import SVC
model = SVC()
```

```python
model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```
